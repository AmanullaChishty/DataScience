```python
import pandas as pd
import urllib.request
from io import StringIO

url = "https://www.openintro.org/data/csv/datasaurus.csv"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "*/*",
    "Referer": "https://www.openintro.org/"
}

req = urllib.request.Request(url, headers=headers)
with urllib.request.urlopen(req) as response:
    csv_data = response.read().decode('utf-8')

datasaurus = pd.read_csv(StringIO(csv_data))
print(datasaurus.head())
```

```python
df = datasaurus.copy()
```

```python
# 1. Descriptive Statistics
print("Descriptive Statistics for 'dino':")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', data=df, s=80, color='teal', edgecolor='black')
plt.title("Scatter Plot - 'dino' Subset")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# 3. Histograms for x and y
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['x'], bins=10, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title("Distribution of x")
sns.histplot(df['y'], bins=10, kde=True, ax=axes[1], color='salmon')
axes[1].set_title("Distribution of y")
plt.show()
```

```python
# 4. Box Plots for x and y
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(y=df['x'], ax=axes[0], color='lightgreen')
axes[0].set_title("Boxplot of x")
sns.boxplot(y=df['y'], ax=axes[1], color='lightyellow')
axes[1].set_title("Boxplot of y")
plt.show()
```
